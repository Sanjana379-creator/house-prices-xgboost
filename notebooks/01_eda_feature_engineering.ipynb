```python
import pandas as pd

df = pd.read_csv('train.csv')
print(df.shape)
print(df.info())
```

```python
print(df['MasVnrType'].value_counts(dropna=False))
```

```python
print(df[df['Fireplaces'] == 0]['FireplaceQu'].isna().sum())
print((df['Fireplaces'] == 0).sum())
```

```python
for col in ['PoolQC', 'Alley', 'Fence', 'FireplaceQu', 'MiscFeature', 'MasVnrType']:
    df[col] = df[col].fillna('None')

print(df[['PoolQC', 'Alley', 'Fence', 'FireplaceQu', 'MiscFeature', 'MasVnrType']].isnull().sum())
```

```python
print(df[df['GarageType'].isna()]['GarageYrBlt'].isna().sum())
print(df['GarageType'].isna().sum())
```

```python
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
```

```python
print(df.groupby('Neighborhood')['LotFrontage'].median().sort_values())
```

```python
df['LotFrontage'] = df.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))
print(df['LotFrontage'].isnull().sum())
```

```python
df['HasGarage'] = df['GarageType'].notna().astype(int)
df['GarageYrBlt'] = df['GarageYrBlt'].fillna(0)
print(df[['GarageYrBlt', 'HasGarage']].isnull().sum())
```

```python
print(df[df['TotalBsmtSF'] == 0][['BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2']].isna().sum())
print((df['TotalBsmtSF'] == 0).sum())
```

```python
print(df[df['GarageType'].isna()][['GarageFinish','GarageQual','GarageCond']].isna().sum())
print(df['Electrical'].isna().sum())
print(df['Electrical'].mode()[0])
```

```python
for col in ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'GarageFinish', 'GarageQual', 'GarageCond']:
    df[col] = df[col].fillna('None')

df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])

print(df.isnull().sum().sum())
```

```python
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
df['GarageType'] = df['GarageType'].fillna('None')
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)

print(df.isnull().sum().sum())
```
